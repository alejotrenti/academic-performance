```python
import pandas as pd

df = pd.read_csv('../data/students.csv')

# Limpieza b√°sica
df.columns = df.columns.str.lower().str.replace(' ', '_')

# Renombrar por comodidad
df.rename(columns={
    'race/ethnicity': 'ethnicity',
    'parental_level_of_education': 'parent_edu',
    'test_preparation_course': 'prep_course'
}, inplace=True)

# Crear promedio general
df['average_score'] = df[['math_score', 'reading_score', 'writing_score']].mean(axis=1)

# Clasificar rendimiento
def classify(score):
    if score >= 90:
        return 'Excelente'
    elif score >= 75:
        return 'Bueno'
    elif score >= 60:
        return 'Regular'
    else:
        return 'Bajo'

df['performance'] = df['average_score'].apply(classify)
df.head()
```
